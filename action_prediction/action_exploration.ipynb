```python
action_explanation = """
This response draws on all the sources you have provided.
There are 35 operations currently available in the O2ARCEnv environment:
Coloring Operations: These operations change the color of selected pixels. There are 10 operations (Color0-Color9) corresponding to the 10 distinct colors defined in ARC.
Flood Fill Operations: These operations change the color of multiple pixels sharing the same color as the selected pixels. Like the coloring operations, there are 10 operations (FloodFill0-FloodFill9) for the 10 colors in ARC.
Object-Oriented Operations: These operations treat the pixels in the grid as objects and perform actions on them. These actions include:
Move: MoveU, MoveD, MoveR, and MoveL move the selected pixels in the specified direction.
Rotate: Rotate90, Rotate180, and Rotate270 rotate the selected pixels by 90, 180, or 270 degrees counterclockwise, respectively.
Flip: FlipH, FlipV, FlipD0, and FlipD1 mirror the selected pixels horizontally, vertically, or diagonally.
Clipboard Operations: These operations use the clipboard to copy and paste pixels:
CopyI: Copies pixels from the test input grid (referred to as input in the state space).
CopyO: Copies pixels from the editing grid (referred to as grid in the state space).
Paste: Pastes the pixels from the clipboard onto the editing grid.
Critical Operations: These operations make significant modifications to the grid:
CopyInput: Replaces the editing grid with the input grid.
ResetGrid: Clears the editing grid by setting all pixels to black.
ResizeGrid: Changes the size of the editing grid.
CropGrid: Crops the editing grid to the selected pixels.
Submit: Submits the current editing grid to be compared to the answer grid.
You can add or remove operations when you subclass the environments. The ARCRawEnv is the simplest environment and only includes the coloring operations and the ResizeGrid and Submit critical operations.
"""

color_description = """
The 10 colors in ARC are:
	1.	#000000 — Black
	2.	#4285F4 — Google Blue or Royal Blue
	3.	#EA4335 — Google Red or Cinnabar Red
	4.	#34A853 — Google Green or Emerald Green
	5.	#FBBC04 — Google Yellow or Golden Yellow
	6.	#9AA0A6 — Slate Gray
	7.	#FF4081 — Hot Pink
	8.	#FF6D00 — Vibrant Orange or Blaze Orange
	9.	#A0D7E7 — Sky Blue or Baby Blue
	10.	#A42E51 — Burgundy or Cranberry
"""

example_action_description = """For example, the action "Color0" can be described as: "The selected pixels are colored black."""
```

```python
client = OpenAI(api_key=api_key)


generated_action_descriptions = {}
for action in new_env.op_names:
    message = f"generate a consise desctiption for the action {action} based on the context information: {action_explanation}. The colornames are defined as: {color_description}. {example_action_description}. The action {action} can be described as:"

    completion = client.chat.completions.create(
        model="gpt-4o",
        messages=[
            {"role": "user", "content": message}
        ]
        )
    description = completion.choices[0].message.content

    response = client.embeddings.create(
        input=description,
        model="text-embedding-3-small"
    )

    embedding = response.data[0].embedding
    generated_action_descriptions[action] = {}
    generated_action_descriptions[action]['description'] = description
    generated_action_descriptions[action]['embedding'] = embedding


print(generated_action_descriptions)

```
